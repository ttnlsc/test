```python
from pathlib import Path
```

```python
table = '../files_for_courses/genomed_run1.csv'
file = open(table)
data = file.read()
file.close()
```

```python
data = data.splitlines()
new_to_old = {}  #очень много старого и нового, старые - это мои, новые - это от тех, кому лениво делать семпл шит

header = data[0]
data = data[1:]

for line in data:
    old, new = line.split(',')
    new = new + '_'
    old = old + '_L001_'
    new_to_old[new] = old
    
new_to_old
```

```python
path = '/media/ttnlsc/08A47ACAA47AB9B0/110823/fastq'

path = Path(path)

for file in path.iterdir():
    name = file.name.split('L00_') # сплит по _ нарезает мой файлик как колбаску
    old_name = name[0]
    file_extension = name[1]
    relevant_name = new_to_old[old_name]
    file.rename(path / f'{relevant_name}{file_extension}') # удивительно, но все получилось
```
